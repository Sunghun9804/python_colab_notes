```python
# 09_test

import random

number = random.randint(1,31) # 1~31 까지의 숫자를 랜덤으로 뽑아서 number에 대입
# print(f'내 마음속 숫자: {number}')

runnung = True

# while 은 오른쪽에 있는 조건 결과가 True 일 경우 반복되는 구문
# runnung 이 초기에 True 이므로 무조건 실행되는 구조

while runnung:
    # 입력받은 값을 정수(int) 로 반환하여 guess 에 대입
    guess = int(input('1~31 중 내가 생각한 숫자는?'))
    print(f'입력받은 숫자: {guess}')
    if number > guess: # number 가 guess 보다 크다면? 아래 내용 출력
        print('내가 생각한 숫자보다 작습니다.')
    elif number < guess: # number 가 guess 보다 작다면? 아래 내용 출력
        print('내가 생각한 숫자보다 큽니다.')
    else: # number 와 guess 가 서로 크지도 작지도 않으면? 아래 내용 출력
        print('정답입니다.')
        runnung = False
```

```python
# 11_list

#파스칼표기법 - 맨 앞글자만 대문자(Java 의 class 생성시 활용)
#카멜표기법 - 단어의 의미가 있는 첫글자를 대문자로 한다.(shopList)
#스네이크표기법 - 단어간의 구분을 _로 한다.(shop_list)

# 리스트 생성 방법 1 - 비어있는 리스트로 생성
a = []

# 리스트 생성 방법2 - 값이이쓴 리스트로 생성
shop_list = ['apple','mango','carrot','banana']
print(f'shop_list : {shop_list}')

# 리스트에 값을 넣는 방법
# 가장뒤로부터 넣는 방법
a.append(1)
a.append(2)
a.append(3)

print(a)

# 특정한 번호를 지정해서 넣는 방법
a.insert(1,'X')
print(a)

print(f'a 의 길이 : {len(a)}') # a 리스트의 길이 = 4
print(f'a: {a}')
# a의 2번방에 있는 값
print(f'a[2]: {a[2]}')
# a의 가장 마지막 방에 있는 값
print(f'a 의 마지막 방에 있는 값: {a[3]}')
# 길이에서 1을 뺀 값을 이용 -> 인덱스는 0번부터 시작한다는 점을 이용
print(f'a 의 마지막 방에 있는 값: {a[len(a)-1]}')
# 파이썬에서 사용되는 방식, 0보다 뒤로가면 맨 뒤로 이동된다는 개념
print(f'a 의 마지막 방에 있는 값: {a[-1]}')

# 리스트 정렬(sort)
shop_list.sort() # 오름차순 -> 작은 수에서 큰 수로 커지는 순서
print(f'shop_list: {shop_list}')

shop_list.sort(reverse=True) # 내림차순 -> 큰 수에서 작은 수로 작아지는 순서
print(f'shop_list: {shop_list}')

# sorted 는 원본의 리스트를 정렬한 값을 새로운 리스트로 반환
new_list = sorted(shop_list)
print(f'new_list: {new_list}')

# a 의 2 번인덱스에 c 를 넣는다
# insert 와 다른점은 해당 인덱스의 값을 지우고 그자리에 들어간다는 것이다.
print(a)
a[2] = 'c'
print(f'a: {a}') # 원래 2번 인덱스에 있던자리 결과 2가 'c'로 바뀜

# list 삭제
print(a)
del a[1]
print(f'a: {a}') # 1번 인덱스 결과 'X'를 삭제
```

```python
# 12_slicing

# 슬라이싱(특정 영역만큼 선택해서 보여주거나 값을 추가하는 방법)
arr = ['Life','is','too','short']
# arr[보여줄시작인덱스:버릴시작인덱스]
area = arr[0:2] # 0~1 까지 보여준다.(0번부터 보여주고 2번부터 버린다.)
print(area)
area = arr[2:4] # 2~3 까지 보여준다.(2번부터 보여주고 4번부터 버린다.)
print(area)
area = arr[:2] # 0번부터 보여주고, 2번부터 버린다.
print(area)
area = arr[2:] # 2번부터 보여주고...
print(area)

# 특정 영역을 지정해서 값을 추가 할 수도 있다.
print(arr)
arr[1:3] = ['A','B','C'] # 1~2 영역을 A,B,C 로 대체 하겠다.
print(arr)
arr[2:3] = ['가','나','다'] # 2 영역을 가,나,다로 대체하겠다.
print(arr)
# 2 영역에 가,나,다 추가
# arr[2:2]은 시작과 끝이 같아 실제로 아무 요소도 선택되지 않는 빈 구간이므로 기존 값은 그대로 두고, 지정한 위치에 새로운 값들을 끼워 넣는 삽입 동작이 일어난다.
arr[2:2] = ['가','나','다']
print(arr)

# 특정 영역을 커서로 드래그 한 다음 수정하느냐, 그냥 클릭만해서 수정하느냐

# 리스트끼리 연산
a = [1,2,3,4,5]
b = ['a','b','c','d']
print(a+b) # list 합치기
print(a*3) # list 반복
```

```python
# 13_search

# 검색 -> 특정한 자료 덩어리에서 원하는 값을 찾는것
a = [3,4,1,2,3,4,'G','F','G']
# 원하는 값의 인덱스 찾기
# 2 라는 값은 어느 위치에 있는가?
print(f'2는 어디?: {a.index(2)}')
print(f'G는 어디?: {a.index('G')}')

# 5번 인덱스 부터 'G'를 찾아라
print(f'G는 어디?: {a.index('G',5)}')
# 값이 없으면 에러(예외)를 발생 시킨다.
# print(a.index('H')) # ValueError: 'H' is not in list

b = [3,4,1,2,3,4,5,6,1,3,2] # 모든 3을 찾아 보세요.

idx = 0
# while True:
#      idx = b.index(3,idx)
#      print(f'3의 값은 {idx}번에 있다.')
#      idx += 1


for n in b: # for in 을 이용하면 list 에 있는 값을 순서대로 하나씩 뽑아낸다.
    if n == 3:
        print(f'3이 있는 인덱스 : {idx}')
    idx += 1

# 리스트 요소 삭제
# del a[3] 과  a.remove(3)
# del 은 특정 인덱스의 값을 지운다.
# remove 는 해당 값을 지운다.(한개만)
print(f'a: {a}')
a.remove(3)
print(f'a: {a}')

# del a[1]
# print(f'a: {a}') # 인덱스 1번 자리 삭제 즉 [1,'X','c',3] => [1,'c',3]

# pop() = append() 의 반대개념
# 맨마지막 요소를 빼낸다.(리스트에서는 사라진다.)
val = a.pop()
print(f'빼낸값: {val}/a:{a}')
val = a.pop()
print(f'빼낸값: {val}/a:{a}')

# 리스트 확장(더하기와 비슷한 개념)
print(f'a: {a}')
print(f'b: {b}')
a.extend(b)
print(a)

# count(val) 특정한 값이 해당 리스트에 몇개 있는지 확인
print(f'a: {a}')
print(f'a 안에 3인 값은 {a.count(3)}개 있다.')
print(f'a 안에 9인 값은 {a.count(9)}개 있다.') # 없는 값은 0을 반환

# a 안에 있는 모든 3을 지워주세요
# a에 있는 3 지우기, 못찾을때 까지? 갯수가 0개 일때 까지?

for i in a:
    if i == 3:
        a.remove(3)

print(a)

while True:
    a.remove(3)
    if a.count(3) == 0:
        break

print(a)
```

```python
# 15_tuple
tu1 = (1,2,3) # tuple 은 []가 아닌 () 로 선언
tu2 = ('a','b','c')
tu3 = (1,2,('a','b'))

print(f'tu1: {tu1}')
print(f'tu2: {tu2}')
print(f'tu3: {tu3}')

# 불러오기
print(f'tu1[1]: {tu1[1]}')
# slicing
print(f'tu2[1:]: {tu2[1:]}')
# 더하기
print(f'tu1+tu2: {tu1+tu2}')
# 곱하기
print(f'tu1*3: {tu1*3}')

# tuple <-- --> list 간의 전환
print(f'tu2: {tu2}')
tu2list = list(tu2)
print(f'{tu2} -> {tu2list}') # () -> []

list2tu = tuple(tu2list)
print(f'{tu2list} -> {list2tu}') # [] -> ()
```

```python
# 16_dictionary
# 사전은 키:값 형태로 되어있다.
# 비슷한 자료구조로는 자바스크립트에 오브젝트, 자바의 맵이 있다.

dic1 ={
    'name':'kim,ji-hoon',
    'phone':'010-1234-1234',
    'age':55
}

dic2 = {
    'name':'hong,gil-dong',
    'phone':'010-2233-5454',
    'friends':['Alice','John','Smith']
}

print(f'dic1: {dic1}')
print(f'dic2: {dic2}')

# 사전에 데이터 넣기 1
a = {'first' : 'a'}
print(f'a: {a}')

#사전에 데이터 넣기 2
a['second'] = 'b'
print(f'a: {a}')

# 사전에서 특정 요소 삭제
del a['second']
print(f'a: {a}')

# 사전의 특정 요소를 꺼내보자(사용법은 List 와 비슷하다.)
print(dic2['name'])
print(dic2['friends'])
# get 메서드를 활용해서도 가져올 수 있다.
print(dic2.get('phone'))
# 특정 키가 없는경우 None 이 아닌 대체 내용으로 반환할수 있음
print(dic2.get('nick', '해당 내용이 없음'))
```

```python
# 17_dic_func

dic = {
    'name':'hong,gil-dong',
    'phone':'010-1234-1234',
    'friends':['Alice','Smith','Jhone']
}

# dic.keys() : 특정한 사전의 키들만 가져와 dict_keys 라는 객체를 반환한다.
print(dic.keys())

for item in dic.keys():
    print(item)

# dict_keys -> list 로 변환
keys = list(dic.keys())
print(f'keys: {keys}')

# dic.values() : 특정 사전의 값만 가져와 dict_values 라는 객체를 반환
print(dic.values())
# list 로 변경해서 values 라는 변수에 담아보자
values = list(dic.values())
print(f'values: {values}')

# dic.items() : 사전의 {키:값} 을 한 쌍으로가져와 dict_items 로 반환한다.
# 각 키와 값은 () 모양으로 보아 tuple 이다.
print(dic.items())
# list 로 변환해 보면 list 안에 각 키와 값이 튜플로 저장되어 있음을 알 수 있다.
li = list(dic.items())
print(li)

# 값을 가져오기(택1)
print(dic)
print(dic.get('name'))
print(dic.get('phone'))

# dic 안에 있는 모든 키와 값을 키:값 형태로 출력해 보자
# 1. 키를 뽑아낸다음, 키를가지고 값을 뽑아내는 방법

for i in dic.keys():
    # print(i)
    print(f'{i} : {dic[i]}')

# 2. 키와 값을 동시에 뽑아낸 다음 거기서 키와 값을 각각 추출하는 방식
for i in dic.items():
    print(f'{i[0]} : {i[1]}')

members = {
    'kim':63, 'lee':88, 'park':97, "gang":77, "hwang":100, "ga":65,
    "na":92, "la":90, "wang":100, "gu":79
}

# 90이상인 사람의 이름만 출력
# print(members)
print(members.items())

for item in members.items():
    if item[1] >=90:
        print(f'이름 : {item[0]}')

# key in dic: 해당 키가 사전에 존재하는지 확인
# 검색 시작여부를 결정할 수 있는 방법
yn = 'kim' in members
print(f'kim이 있는가? {yn}')

yn = 'jung' in members
print(f'jung이 있는가? {yn}')

# update : 이미 있는 키면 수정을, 없는 키면 추가를 하는 함수
dic.update({'name':'홍길동','age':30, 'married':False})
print(dic)

# dic.clear() : 사전안의 내용을 모두 지운다.
dic.clear()
print(dic)
```

```python
# 18_set

# set 은 순서가 없고, 중복을 허용하지 않는다.
number_set = set([1,2,3])
print(number_set)

# 중복을 제외하고 담는다.
# 그래서 중복 제거용도로 사용된다.
str_set = set("HelloWorld")
print(str_set)

# set 들을 이용해서 집합을 구현할 수 있다.
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])
print(s1)
print(s2)

# 교집합(intersection)
print(f'교집합1: {s1&s2}')
print(f'교집합2: {s1.intersection(s2)}')

#합집합(union)
print(f'합집합1: {s1|s2}')
print(f'합집합2: {s1.union(s2)}')

#차집합(minus|difference)
print(f'차집합1: {s1-s2}')
print(f'차집합2: {s2-s1}')

# 값 1개 추가하기
print(s1)
s1.add(10)
print(s1)

# 여러개 추가하기
print(s1)
s1.update([10,20,30])
print(s1)

# 특정값 제거
s1.remove(10)
print(s1)
# s1.remove(10) <- 없는 값을 지우려고 하면 KeyError 발생
s1.discard(10) # remove 와 같으나 없는 값을 지우려고 해도 에러가 나지 않는다.
```
